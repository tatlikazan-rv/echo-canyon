# Char level generation (only done for pop)

```python
# insert links of lyricsfreak artist pages

artist_links = ["https://www.lyricsfreak.com/b/beatles/",
                "https://www.lyricsfreak.com/m/madonna/",
                "https://www.lyricsfreak.com/e/elton+john/",
                "https://www.lyricsfreak.com/m/michael+jackson/",
                "https://www.lyricsfreak.com/s/rihanna/"]

# helper functions
def get_links(artist_link):
    page = requests.get(artist_link)
    soup = BeautifulSoup(page.content, 'html.parser')
    a = soup.findAll("a", {"class": "lf-link lf-link--primary"})
    links = [i["href"] for i in a]
    print(str(len(links)) + " lyrics found at " + artist_link)
    return(links)

def get_lyrics(links):
    lyrics = {}
    for l in links: 
        page = requests.get("https://www.lyricsfreak.com" + str(l))
        soup = BeautifulSoup(page.content, 'html.parser')
        title = re.sub("\+", "_", l.split("/")[-1].split("_")[0])
        lyrics[title] = soup.find(id = "content")
    return(lyrics)

# some strings in the lyrics that will be removed
beg = re.compile('<div class.*')
end = re.compile('</div>')
lin = re.compile('<br/>')
cho = re.compile('chorus|verse|intro', flags = re.IGNORECASE)
cr = re.compile('\r')
patterns = [beg, end, lin, cho, cr]

def clean_text(patterns, text):
    text = str(text)
    for pattern in patterns:
        text = re.sub(pattern, "", text)
    return(text.strip())
        
lyrics_dict = {}

save_path = "/content/drive/MyDrive/EchoCanyon/lyrics/"

def reverse_str(x):
    return x[::-1]

for artist_link in artist_links:
    
    lyrics_links = [link for link in (get_links(artist_link))]
    lyrics = get_lyrics(lyrics_links)

    lyrics = {title: reverse_str(clean_text(patterns,lyric)) for title, lyric in lyrics.items()}

    lyrics_dict[re.sub("\+", "_", artist_link.split("/")[-2])] = lyrics

with open(save_path + "pop_lyrics_rev.json", "w", encoding = "utf8") as f:
    
    json.dump(lyrics_dict, f)

#lyrics_dict
```

```python
# specifying a set of characters for the model
vocabulary_def = re.findall(r"[ 0-9a-zA-Z\n.,!?']", string.printable)
print(vocabulary_def)

lyrics_dict = lyrics_dict_rev
```

```python
# translation to integers and back

char_to_int = preprocessing.StringLookup(
    vocabulary=list(vocabulary_def))

int_to_char = preprocessing.StringLookup(
    vocabulary=char_to_int.get_vocabulary(), invert=True)

def int_to_text(integers):
    return tf.strings.reduce_join(int_to_char(integers), axis=-1)

example = ["H", "e", "l", "l", "o", " ", "&"]
print(example)
print(char_to_int(example))
print(int_to_char(char_to_int(example)))
print(int_to_text(char_to_int(example)))

# Transforming to tokens and back gives non defined (not in our vocabulary) characters the label [UNKNOWN]
```

```python
# splitting our lyrics in chunks of equal length
dataset = None
seq_length = 100 # = number of recurrent operations

for artist, titles in lyrics_dict.items():
    
    for title, lyric in titles.items():
        
        lyric = tf.strings.unicode_split(lyric, input_encoding='UTF-8')
        lyric = char_to_int(lyric)
        lyric = tf.data.Dataset.from_tensor_slices(lyric)
        lyric = lyric.batch(seq_length+1, drop_remainder=True)  
        # we need an output for the last input then the next 
        dataset = lyric if dataset is None else dataset.concatenate(lyric)
        
[print(lyric, int_to_text(lyric.numpy())) for lyric in dataset.take(1)];
```

```python
# inputs and targets stem from the same character sequence, targets are offset by one character
def split_input_target(sequence):
    input_text = sequence[:-1]
    target_text = sequence[1:]
    return input_text, target_text

dataset = dataset.map(split_input_target)

[print(int_to_text(input_), "\n", int_to_text(target)) for input_, target in dataset.take(1)];
```

```python
# shuffling and batching
buffer_size = len(dataset)
batch_size = 64
dataset = dataset.shuffle(buffer_size).batch(batch_size, drop_remainder = True)

# for efficiency
dataset = dataset.cache().prefetch(tf.data.experimental.AUTOTUNE)

dataset # (batchsize, sequence length) as input and output
```

```python
# subclassing tf.keras.Model
class OurModel(tf.keras.Model):
    
    def __init__(self, vocab_size, embedding_dim, hidden_dim):  
        super().__init__(self)

        # here we define our layers:
        self.embedding = tf.keras.layers.Embedding(vocab_size, embedding_dim)
        self.LSTM = tf.keras.layers.LSTM(hidden_dim,
                                       return_sequences=True,
                                       return_state=True)
        self.dense = tf.keras.layers.Dense(vocab_size)

    def call(self, inputs, hidden_state = None, cell_state = None, return_state=False, training=False):
        # Training param is to stop the training when we are calling prediction
        x = inputs
        x = self.embedding(x, training=training)
        if hidden_state is None:
            hidden_state, cell_state = self.LSTM.get_initial_state(x)
        x, hidden_state, cell_state = self.LSTM(x, initial_state=[hidden_state, cell_state], training=training)
        x = self.dense(x, training=training)

        if return_state:
            return x, hidden_state, cell_state
        else:
            return x
```

```python
# What should our dimensions be?
vocab_size = char_to_int.vocabulary_size() # = input_dim
embedding_dim = 75  # this is a hyperparameter, it basically turns one char to a vector sized embedding_dim
hidden_dim = 128 # this is also a hyperparameter
# we dont actually need one but,output_dim = char_to_int.vocabulary_size() # as we are predicting characters
```

```python
model = OurModel(
    vocab_size=vocab_size,
    embedding_dim=embedding_dim,
    hidden_dim=hidden_dim)
```

```python
class TextGenerator:
    def __init__(self, model, char_to_int, int_to_char):
        self.model = model
        self.char_to_int = char_to_int
        self.int_to_char = int_to_char
        
        sparse_mask = tf.SparseTensor(
            values=[-float('inf')],
            indices=char_to_int(['[UNK]'])[:, None],
            dense_shape=[char_to_int.vocabulary_size()])
        self.prediction_mask = tf.sparse.to_dense(sparse_mask) 
        # we wnat to supress the prediction as [UNK] tokens, set their probability to negative infinity  
        
    def generate_char(self, inputs, hidden_state = None, cell_state = None):
        
        # processing inputs
        chars = tf.strings.unicode_split(inputs, 'UTF-8')
        integers = tf.expand_dims(self.char_to_int(chars), axis = 0)
    
        # running the model
        x, hidden_state, cell_state = self.model(inputs=integers, 
                                                      hidden_state= hidden_state, cell_state = cell_state,
                                                      return_state=True)
        
        logits = x[:,-1,:]  # we want to predict last output after 100 sequences 
        logits = logits + self.prediction_mask
        
        # sampling
        temperature = 0.5 # the temperature controls the randomness of the predictions
        # high tempreture gives back the random predictions we had before training
        # low temprature gives back repetitions
        preds = tf.random.categorical(logits/temperature, num_samples=1)
        
        preds = tf.squeeze(preds, axis=-1)
    
        # translating to a character
        preds = self.int_to_char(preds)
    
        return preds, hidden_state, cell_state
    
    
    def generate_text(self, inputs, n_characters = 200, hidden_state= None, cell_state = None):
    
        outputs = ""
    
        for i in range(0, n_characters):
        
            char, hidden_state, cell_state = generator.generate_char(inputs, hidden_state = hidden_state, cell_state = cell_state, )
        
            outputs = outputs + char.numpy()[0].decode("utf-8")
            inputs = (inputs + outputs)[-200:]
        
        return outputs
        
```
